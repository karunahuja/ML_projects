```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
arg2=StandardScaler()
titanic_data=pd.read_csv('Datasets/titanic_train.csv')
#print(titanic_data)
def f(str):
    if str=="S":
        return 0
    elif str=='c':
        return 1
    else:
        return 2

def fu(s):
    if s<=16:
        return 0
    elif s>16 and s<=32:
        return 1
    elif s>32 and s<=48:
        return 2
    elif s>48 and s<=64:
        return 3
    else:
        return 4
        
def fus(s):
    if 'Master' in s:
        return 0
    elif 'Miss' in s:
        return 1
    elif 'Mr' in s:
        return 2
    elif 'Mrs' in s:
        return 3
    elif 'Royalty' in s:
        return 4
    else:
        return 5

x_train=titanic_data.copy()

x_train['class']=x_train.Name.apply(fus)
x_train.drop('Name',inplace=True,axis=1)
x_train['Sex']=x_train['Sex'].map({'female':1,'male':0}).astype(int)
x_train.drop('Ticket',inplace=True,axis=1)
x_train.drop('Cabin',inplace=True,axis=1)
x_train.drop('Fare',inplace=True,axis=1)
x_train['family']=x_train['SibSp']+x_train['Parch']+1
del x_train['SibSp']
del x_train['Parch']
x_train['Embarke']=x_train.Embarked.apply(f)
x_train.drop('Embarked',inplace=True,axis=1)

x_train.drop('Survived',inplace=True,axis=1)
x_train.Age.fillna(x_train.Age.mean(),inplace=True)
x_train['new_age']=x_train.Age.apply(fu)
del x_train['Age']
xtrain=np.array(x_train)



ytrain=titanic_data.Survived
df=pd.DataFrame(xtrain)

#print(xtrain)

features=df.columns
b=6
for a in features:
    df[b]=df[a]*df[a]
    b+=1
c=12
for a in features:
    df[c]=df[a]*df[a]*df[a]
    c+=1

d=18
for i in range(6):
    for j in range(i+1,6):
        df[d]=df[features[i]]*df[features[j]]
        d+=1
xtrain=np.array(df)






xtrain=arg2.fit_transform(xtrain)





arg=LogisticRegression()
arg.fit(xtrain,ytrain)
x_test=pd.read_csv('Datasets/titanic_test.csv')

x_test['class']=x_test.Name.apply(fus)
x_test.drop('Name',inplace=True,axis=1)
x_test['Sex']=x_test['Sex'].map({'female':1,'male':0}).astype(int)
x_test.drop('Ticket',inplace=True,axis=1)
x_test.drop('Cabin',inplace=True,axis=1)

x_test.drop('Fare',inplace=True,axis=1)
x_test['family']=x_test['SibSp']+x_test['Parch']+1
del x_test['SibSp']
del x_test['Parch']
x_test['Embarke']=x_test.Embarked.apply(f)
x_test.drop('Embarked',inplace=True,axis=1)

x_test.Age.fillna(x_test.Age.mean(),inplace=True)
x_test['new_age']=x_test.Age.apply(fu)
del x_test['Age']
xtest=np.array(x_test)

df=pd.DataFrame(xtest)

features=df.columns
b=6
for a in features:
    df[b]=df[a]*df[a]
    b+=1
c=12
for a in features:
    df[c]=df[a]*df[a]*df[a]
    c+=1
    
    
    
    
x_test=np.array(df)
x_test=arg2.transform(x_test)

pred=arg.predict(x_test)
#print(x_test)
np.savetxt('titanic_predict.csv',pred,delimiter=',')
```
